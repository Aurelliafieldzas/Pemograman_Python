```
import numpy as np
import matplotlib.pyplot as plt
```

```
ntl = lambda x: 10*np.log10(x)
ltn = lambda x: 10**(x/10)

def cos(frek,tipe,ht,hr,d):
    if frek in range(150,1501):
        Ca=69.55
        Cb=26.16
    elif frek in range(1501,2001):
        Ca=46.3
        Cb=33.9
    
    if tipe=='urban' :
        Cm=0
    elif tipe=='suburban' :
        Cm= -2*((np.log10((frek/28)**2))-5.4)
    elif tipe=='open' :
        Cm= -4.78*(np.log10(frek))+18.33*(np.log10(frek))-40.94
    else :
        Cm='tipe salah'
    
    ahr = (1.1*np.log10(frek)-0.7)*hr-(1.56*np.log10(frek)-1.8)
    Lp = Ca+(Cb*np.log10(frek))-(13.83*np.log(ht))-ahr+((44.9-6.55*np.log10(ht))*np.log10(d))-Cm
    
    return Lp
```

```
ht = 80
hr = 1.5
dmin=300 
dmax=1000 
N = 200
ptr = 30
gtr = 10
gre = 2
bw = 5000
d=np.random.randint(dmin,dmax,N)
```

```
#fungsi daya terima, snr, kapasitas
def dterima(ptr,pl,gtr,gre):
  ptr1 = ntl(ptr)
  daya = ptr - pl + gtr + gre
  daya1 = ltn(daya)
  return daya1

def snr(dterima,bw):
  noise = -174 + 10*np.log10(bw)
  noise1 = ltn(noise)
  snr = dterima/noise1
  return snr

def kselu(snr,bw):
  snr1 = ltn(snr)
  kselu = bw*np.log2(1+snr1)
  return kselu
```

```
#daya terima f sama
def dterimaurban():
  f=1800 
  loc='urban'
  pl=cos(f,loc,ht,hr,d)
  dterimafix = dterima(ptr,pl,gtr,gre)
  d_sort=d[np.argsort(d)]
  Urban_sort=dterimafix[np.argsort(d)]
  return d_sort,Urban_sort

def dterimaopen():
  f=1800 
  loc='open'
  pl=cos(f,loc,ht,hr,d)
  dterimafix = dterima(ptr,pl,gtr,gre)
  d_sort=d[np.argsort(d)]
  Urban_sort=dterimafix[np.argsort(d)]
  return d_sort,Urban_sort

def dterimasuburban():
  f=1800 
  loc='suburban'
  pl=cos(f,loc,ht,hr,d)
  dterimafix = dterima(ptr,pl,gtr,gre)
  d_sort=d[np.argsort(d)]
  Urban_sort=dterimafix[np.argsort(d)]
  return d_sort,Urban_sort
  
d11,pl11 = dterimaurban()
d12,pl12 = dterimaopen()
d13,pl13 = dterimasuburban()


#daya terima lokasi sama
def dterima700():
  f=700
  loc='urban'
  pl=cos(f,loc,ht,hr,d)
  dterimafix = dterima(ptr,pl,gtr,gre)
  d_sort=d[np.argsort(d)]
  Urban_sort=dterimafix[np.argsort(d)]
  return d_sort,Urban_sort

def dterima900():
  f=900
  loc='urban'
  pl=cos(f,loc,ht,hr,d)
  dterimafix = dterima(ptr,pl,gtr,gre)
  d_sort=d[np.argsort(d)]
  Urban_sort=dterimafix[np.argsort(d)]
  return d_sort,Urban_sort

def dterima1800():
  f=1800 
  loc='urban'
  pl=cos(f,loc,ht,hr,d)
  dterimafix = dterima(ptr,pl,gtr,gre)
  d_sort=d[np.argsort(d)]
  Urban_sort=dterimafix[np.argsort(d)]
  return d_sort,Urban_sort

d21,pl21 = dterima700()
d22,pl22 = dterima900()
d23,pl23 = dterima1800()
```

```
plt.figure()
plt.plot(d11,pl11,'r', label='urban')
plt.plot(d12,pl12,'g', label='open')
plt.plot(d13,pl13,'b', label='suburban')
plt.legend()
plt.title('Daya Terima')
plt.xlabel('Jarak (m)')
plt.ylabel('Daya Terima (miliwatt)')
plt.show()

plt.figure()
plt.plot(d21,pl21,'r', label='700 Mhz')
plt.plot(d22,pl22,'g', label='900 Mhz')
plt.plot(d23,pl23,'b', label='1800 Mhz')
plt.legend()
plt.title('Daya Terima')
plt.xlabel('Jarak (m)')
plt.ylabel('Daya Terima (miliwatt)')
plt.show()
```
